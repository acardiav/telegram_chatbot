```python
import os
os.environ['DJANGO_SETTINGS_MODULE']= 'telegram_chatbot.settings'
os.environ['DJANGO_ALLOW_ASYNC_UNSAFE']= 'true'

import django
django.setup()

from urllib.parse import urlparse
import requests
from bs4 import BeautifulSoup
import re
```

```python
req = requests.get("http://sj.hongik.ac.kr/site/food/food_menu.html")
html = req.content
soup = BeautifulSoup(html, 'lxml') # pip install lxml
print(soup)
```

```python
list_day = ['월요일','화요일','수요일','목요일','금요일','토요일']

list_food = soup.find_all(name="div", attrs={"class":"foodmenu"})
print(list_food)
```

```python
crawlingmenu = []
for index in range(0, len(list_day)):
    title = list_food[index].find().text
    day = list_day[index]
    obj = {
        'title' : title,
        'day' : day
    }
    crawlingmenu.append(obj)
    print(day,'\n',title.lstrip())
```

```python
from crawling.models import Menu
```

```python
menu_list = []

for menu_dict in crawlingmenu:
    menu = Menu(title=menu_dict['title'], day=menu_dict['day'])
    menu_list.append(menu)
    
menu_list
Menu.objects.bulk_create(menu_list)
```
